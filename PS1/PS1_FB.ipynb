```python
# 计算所有子集平均值的总和
def Sum_averages(arr):
    n = len(arr)
    total_sum = 0
    
    # 遍历所有可能的子集（不包括空集）
    for i in range(1, 1 << n):  # 从1开始，跳过空集
        subset = []
        # 检查每一位，决定哪些元素在子集中
        for j in range(n):
            if i & (1 << j):  # 如果第j位是1，则包含该元素
                subset.append(arr[j])
        
        # 计算当前子集的平均值并累加
        avg = sum(subset) / len(subset)
        total_sum += avg
    
    return total_sum

def Sum_averages_fast(arr):
    """使用数学公式 O(N) 时间复杂度"""
    n = len(arr)
    if n == 0:
        return 0
    total_sum = sum(arr)
    return total_sum * (2**n-1) / n

# 测试题目
print("题目4.2测试:")
test_arr = [1, 2, 3]
result = Sum_averages_fast(test_arr)
print(f"数组 {test_arr} 的所有子集平均值总和: {result}")
```

```python
import random
import matplotlib.pyplot as plt

# 先定义前两个函数
def Random_integer(N):
    return [random.randint(0, 10) for _ in range(N)]

def Sum_averages(arr):
    n = len(arr)
    total_sum = 0
    for i in range(1, 1 << n):
        subset = []
        for j in range(n):
            if i & (1 << j):
                subset.append(arr[j])
        avg = sum(subset) / len(subset)
        total_sum += avg
    return total_sum

# 4.3 测试和绘图
Total_sum_averages = []

for N in range(1, 101):  # 计算1到20，避免太慢
    arr = Random_integer(N)
    result = Sum_averages_fast(arr)
    Total_sum_averages.append(result)
    print(f"N={N}: {result:.2f}")

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(range(1, 101), Total_sum_averages, 'bo-', linewidth=2)
plt.xlabel('Array Size (N)')
plt.ylabel('Sum of Averages')
plt.title('Sum of Averages of All Subsets vs Array Size')
plt.grid(True, alpha=0.3)
plt.show()

print("观察：曲线呈指数增长，因为子集数量是2^N")
```

```python
import random

# 5.1 创建矩阵
def create_matrix(N, M):
    """
    创建N行M列的矩阵
    左上角和右下角固定为1，其他位置随机为0或1
    """
    # 创建全0矩阵
    matrix = [[0 for _ in range(M)] for _ in range(N)]
    
    # 设置左上角和右下角为1
    matrix[0][0] = 1
    matrix[N-1][M-1] = 1
    
    # 随机填充其他位置
    for i in range(N):
        for j in range(M):
            # 跳过左上角和右下角
            if (i == 0 and j == 0) or (i == N-1 and j == M-1):
                continue
            matrix[i][j] = random.randint(0, 1)
    
    return matrix

# 5.2 计算路径数
def Count_path(matrix):
    """
    计算从左上角到右下角的路径总数
    只能向右或向下移动，0表示障碍物
    """
    N = len(matrix)
    M = len(matrix[0])
    
    # 创建DP表，dp[i][j]表示到达(i,j)的路径数
    dp = [[0 for _ in range(M)] for _ in range(N)]
    
    # 初始化起点
    if matrix[0][0] == 1:
        dp[0][0] = 1
    else:
        return 0  # 如果起点就是障碍，没有路径
    
    # 填充第一行：只能从左边来
    for j in range(1, M):
        if matrix[0][j] == 1:
            dp[0][j] = dp[0][j-1]  # 只能从左边来
        else:
            dp[0][j] = 0  # 障碍物，无法到达
    
    # 填充第一列：只能从上面来
    for i in range(1, N):
        if matrix[i][0] == 1:
            dp[i][0] = dp[i-1][0]  # 只能从上面来
        else:
            dp[i][0] = 0  # 障碍物，无法到达
    
    # 填充其他位置
    for i in range(1, N):
        for j in range(1, M):
            if matrix[i][j] == 1:
                # 可以从上面或左边来
                dp[i][j] = dp[i-1][j] + dp[i][j-1]
            else:
                dp[i][j] = 0  # 障碍物，无法到达
    
    return dp[N-1][M-1]

# 5.3 运行1000次并计算平均值
def run_experiment():
    N = 10
    M = 8
    total_paths = 0
    num_runs = 1000
    
    print(f"开始实验: N={N}, M={M}, 运行{num_runs}次")
    
    for run in range(num_runs):
        # 创建新矩阵
        matrix = create_matrix(N, M)
        
        # 计算路径数
        paths = Count_path(matrix)
        total_paths += paths
        
        # 每100次显示一次进度
        if (run + 1) % 100 == 0:
            print(f"已完成 {run + 1} 次运行，当前平均路径数: {total_paths / (run + 1):.2f}")
    
    mean_paths = total_paths / num_runs
    print(f"\n最终结果:")
    print(f"1000次运行的平均路径数: {mean_paths:.2f}")
    
    return mean_paths

# 测试小例子
def test_small_example():
    print("测试小例子:")
    # 创建一个简单的3x3矩阵测试
    test_matrix = [
        [1, 1, 1],
        [1, 0, 1],
        [1, 1, 1]
    ]
    
    print("测试矩阵:")
    for row in test_matrix:
        print(row)
    
    paths = Count_path(test_matrix)
    print(f"路径数: {paths}")

# 主程序
if __name__ == "__main__":
    # 运行测试例子
    test_small_example()
    print("\n" + "="*50 + "\n")
    
    # 运行主要实验
    mean_result = run_experiment()
```
